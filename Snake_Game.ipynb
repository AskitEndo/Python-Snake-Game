```python
import turtle
import time
import random

delay = 0.1
score = 0
high_score = 0

# Creating a window screen
wn = turtle.Screen()
wn.title("Snake Game")
wn.bgcolor("#121212")  # Dark background
wn.setup(width=600, height=600)
wn.tracer(0)

# Head of the snake
head = turtle.Turtle()
head.shape("triangle")  # Change the shape to triangle
head.color("#F5F5F5")  # Off-white color
head.penup()
head.goto(0, 0)
head.direction = "Stop"

# Food in the game
food = turtle.Turtle()
food_speed = 0
food.shape("circle")
food.color("#6B8E23")  # Olive green color
food.penup()
food.goto(0, 100)

# Fppf (hazard) in the game
fppf = turtle.Turtle()
fppf.speed(0)
fppf.shape("circle")
fppf.color("red")
fppf.penup()
fppf.goto(0, 0)
fppf.hideturtle()

pen = turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color("#F5F5F5")  # Off-white color
pen.penup()
pen.hideturtle()
pen.goto(0, 250)
pen.write("Score: 0 High Score: 0", align="center", font=("Arial", 24, "bold"))

# Texts at the bottom corners
bottom_text_font = ("Arial", 10, "normal")

# Python Snake Game
game_text = turtle.Turtle()
game_text.speed(0)
game_text.color("#F5F5F5")  # White color
game_text.penup()
game_text.hideturtle()
game_text.goto(-280, -280)
game_text.write("Python3 Snake Game", align="left", font=bottom_text_font)

# "by AskitEndo"
author_text = turtle.Turtle()
author_text.speed(0)
author_text.color("#F5F5F5")  # White color
author_text.penup()
author_text.hideturtle()
author_text.goto(280, -280)
author_text.write("by AskitEndo", align="right", font=bottom_text_font)

# Assigning key directions
def go_up():
    if head.direction != "down":
        head.direction = "up"
        head.setheading(90)  # Pointing up

def go_down():
    if head.direction != "up":
        head.direction = "down"
        head.setheading(270)  # Pointing down

def go_left():
    if head.direction != "right":
        head.direction = "left"
        head.setheading(180)  # Pointing left

def go_right():
    if head.direction != "left":
        head.direction = "right"
        head.setheading(0)  # Pointing right

def move():
    if head.direction == "up":
        y = head.ycor()
        head.sety(y + 20)
    if head.direction == "down":
        y = head.ycor()
        head.sety(y - 20)
    if head.direction == "left":
        x = head.xcor()
        head.setx(x - 20)
    if head.direction == "right":
        x = head.xcor()
        head.setx(x + 20)

# Keyboard bindings
wn.listen()
wn.onkeypress(go_up, "w")
wn.onkeypress(go_down, "s")
wn.onkeypress(go_left, "a")
wn.onkeypress(go_right, "d")

segments = []

# Main Gameplay
while True:
    wn.update()

    # Border collision detection
    if head.xcor() > 290 or head.xcor() < -290 or head.ycor() > 290 or head.ycor() < -290:
        time.sleep(1)
        head.goto(0, 0)
        head.direction = "Stop"
        colors = random.choice(['red', 'blue', 'green'])
        shapes = random.choice(['square', 'circle'])
        for segment in segments:
            segment.goto(1000, 1000)
        segments.clear()
        score = 0
        delay = 0.1
        pen.clear()
        pen.write("Score: {} High Score: {} ".format(score, high_score), align="center", font=("Arial", 24, "bold"))

    # Collision with food
    if head.distance(food) < 20:
        x = random.randint(-270, 270)
        y = random.randint(-270, 270)
        food.goto(x, y)

        # Adding segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("orange")  # Tail color
        new_segment.penup()
        segments.append(new_segment)
        delay -= 0.001
        score += 10
        if score > high_score:
            high_score = score
        pen.clear()
        pen.write("Score: {} High Score: {} ".format(score, high_score), align="center", font=("Arial", 24, "bold"))

    # Collision with fppf (hazard)
    if head.distance(fppf) < 20:
        time.sleep(1)
        head.goto(0, 0)
        head.direction = "Stop"
        for segment in segments:
            segment.goto(1000, 1000)
        segments.clear()
        score = 0
        delay = 0.1
        pen.clear()
        pen.write("Score: {} High Score: {} ".format(score, high_score), align="center", font=("Arial", 24, "bold"))

    # Move the end segments first in reverse order
    for index in range(len(segments) - 1, 0, -1):
        x = segments[index - 1].xcor()
        y = segments[index - 1].ycor()
        segments[index].goto(x, y)

    # Move segment 0 to where the head is
    if len(segments) > 0:
        x = head.xcor()
        y = head.ycor()
        segments[0].goto(x, y)

    move()

    # Check for head collisions with the body segments
    for segment in segments:
        if segment.distance(head) < 20:
            time.sleep(1)
            head.goto(0, 0)
            head.direction = "Stop"
            for segment in segments:
                segment.goto(1000, 1000)
            segments.clear()
            score = 0
            delay = 0.1
            pen.clear()
            pen.write("Score: {} High Score: {} ".format(score, high_score), align="center", font=("Arial", 24, "bold"))

    # Generate fppf (hazard) randomly
    if random.randint(1, 100) == 1:
        fppf.goto(random.randint(-270, 270), random.randint(-270, 270))
        fppf.showturtle()

    time.sleep(delay)

wn.mainloop()
```
